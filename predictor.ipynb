```python
import pandas as pd
import numpy as np
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from autots import AutoTS
import matplotlib.pyplot as plt

# === Step 1: Download BTC-USD Data ===
ticker = "BTC-USD"
df = yf.download(ticker, start="2023-08-10", end="2024-08-10")

# === Step 2: Reset index and flatten columns ===
df.reset_index(inplace=True)
df.columns = [col[0].lower() if isinstance(col, tuple) else col.lower() for col in df.columns]

# === Step 3: Rename columns ===
rename_map = {
    "date": "date",
    "close": "close",
    "high": "high",
    "low": "low",
    "open": "open",
    "volume": "volume"
}
df.rename(columns=rename_map, inplace=True)
print("Columns after rename:", df.columns.tolist())

# === Step 4: Scale 'close' column ===
scaler = MinMaxScaler()
df['close'] = scaler.fit_transform(df[['close']])
print(df[['date', 'close']].head())

# === Step 5: Prepare dataframe for AutoTS ===
df_autots = df[['date', 'close']].copy()
df_autots['date'] = pd.to_datetime(df_autots['date'])

# === Step 6: Initialize AutoTS Model ===
model = AutoTS(
    forecast_length=30,
    frequency='D',
    ensemble='simple',
    model_list="superfast",  # faster; use 'default' or 'all' for more accuracy
    transformer_list='fast',
    max_generations=5,
    num_validations=2
)

# === Step 7: Fit Model ===
model = model.fit(
    df_autots,
    date_col='date',
    value_col='close',
    id_col=None
)

# === Step 8: Generate Forecast ===
prediction = model.predict()
forecast = prediction.forecast

# === Step 9: Inverse transform predictions ===
forecast_inv = scaler.inverse_transform(forecast)
forecast_df = pd.DataFrame({
    'date': forecast.index,
    'forecast_btc_usd': forecast_inv.flatten()
})

# === Step 10: Save forecast to CSV ===
forecast_df.to_csv("btc_forecast.csv", index=False)
print("Forecast saved to btc_forecast.csv")

# === Step 11: Plot ===
plt.figure(figsize=(12, 6))
plt.plot(df_autots['date'], scaler.inverse_transform(df_autots[['close']]), label="Historical")
plt.plot(forecast_df['date'], forecast_df['forecast_btc_usd'], label="Forecast", linestyle="--")
plt.xlabel("Date")
plt.ylabel("BTC Price (USD)")
plt.title("BTC-USD Price Forecast (AutoTS)")
plt.legend()
plt.show()
```
